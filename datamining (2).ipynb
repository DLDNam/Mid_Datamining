```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score
```

```python
X = df.drop(['Loan-On-Card'], axis=1)

y = df['Loan-On-Card']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.shape, X_test.shape
```

```python
cols = X_train.columns
```

```python
# khi test 
# y_test = ....
# X_test = ....
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=[cols])
```

```python
X_test = pd.DataFrame(X_test, columns=[cols])
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state =42)
X_train, y_train = smote.fit_resample(X_train,y_train)
```

```python
X_train.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)


# fit the model to the training set
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.4f}'.format(accuracy_score(y_test, y_pred)))
```

```python
#check overfitting
print('Training set score: {:.4f}'.format(knn.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(knn.score(X_test, y_test)))
```

```python
#Rebuild KNN classification model using 
for k in range(5,20):
    knn_5 = KNeighborsClassifier(n_neighbors=k)


    # fit the model to the training set
    knn_5.fit(X_train, y_train)

    # predict on the test-set
    y_pred_5 = knn_5.predict(X_test)


    print('Model accuracy score with k=', k ,' : {0:0.4f}'. format(accuracy_score(y_test, y_pred_5)))
```

```python
#Rebuild KNN classification model using k=10
knn_10 = KNeighborsClassifier(n_neighbors=10)
# fit the model to the training set
knn_10.fit(X_train, y_train)
# predict on the test-set
y_pred_10 = knn_10.predict(X_test)

print('Model accuracy score with k=11 : {0:0.4f}'. format(accuracy_score(y_test, y_pred_10)))
```

```python
#confusion matrix
# Print the Confusion Matrix with k =3 and slice it into four pieces

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_10)

print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
# visualize confusion matrix with seaborn heatmap

plt.figure(figsize=(6,4))

cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_10))
```

```python
# Applying 10-Fold Cross Validation

from sklearn.model_selection import cross_val_score

scores = cross_val_score(knn_10, X_train, y_train, cv = 10, scoring='accuracy')

print('Cross-validation scores:{}'.format(scores))
```

```python
    # compute Average cross-validation score

print('Average cross-validation score: {:.4f}'.format(scores.mean()))
```
